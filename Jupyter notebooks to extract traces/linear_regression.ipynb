```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
```

```python
df.info()
```

```python
X = df[['GrLivArea']].to_numpy()
Y = df['SalePrice'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
training_iterations = 10000
learning_rate = 0.0000000001
```

```python
def compute_cost(y_hat, y):
  mse = 0
  for i in range(len(y)):
    mse = mse + (y[i]-y_hat[i])*(y[i]-y_hat[i])
  mse = mse/(2*len(y))
  return mse
```

```python
def compute_gradients(y_hat, y, x):
  dj_dw=0
  dj_db=0
  for i in range(len(y)):
    dj_dw = dj_dw + (y_hat[i]-y[i])*x[i]/len(y)
    dj_db = dj_db + (y_hat[i]-y[i])/len(y)

  return dj_dw , dj_db
```

```python
cost_values = []
predictions = np.zeros(len(X))

w = np.random.rand()
b = np.random.rand()

for _ in range(training_iterations):
  for i in range(len(X_train)):
    pred = w*X_train[i] + b
    predictions[i] = pred
  cost_values.append(compute_cost(predictions, y_train))
  dj_dw,dj_db = compute_gradients(predictions,y_train,X_train)
  (w,b) = (w - (learning_rate * dj_dw),b - (learning_rate * dj_db))
```

```python
w,b
```

```python
plt.figure(figsize=(6, 6), dpi=200)
plt.plot(cost_values)
plt.ylabel("cost value")
plt.xlabel("training iteration");
```

```python
plt.figure(figsize=(15, 8), dpi=200)
plt.scatter(X_test, y_test)
plt.plot(X_test,(w*X_test + b), color='red')
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice");
```
